```python
# Calcular Q1 e Q3

Q1 = base['RENDAMENSAL'].quantile(0.25)

Q3 = base['RENDAMENSAL'].quantile(0.75)

 

# Calcular o IQR

IQR = Q3 - Q1

 

# Definir os limites para outliers

limite_inf = Q1 - 1.5 * IQR

limite_sup = Q3 + 1.5 * IQR




```
